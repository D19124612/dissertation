```python
from scipy.stats import shapiro
from scipy.stats import ttest_rel
import numpy as np
import pandas as pd
from matplotlib import pyplot
from statsmodels.graphics.gofplots import qqplot

```

```python
bert_results = pd.read_csv('/content/drive/MyDrive/Dissertation/results/bert_baselines.csv')
```

```python
distilbert_results = pd.read_csv('/content/drive/MyDrive/Dissertation/results/distilbert_baselines.csv')
```

```python
distilbert_llrd = pd.read_csv('/content/drive/MyDrive/Dissertation/results/distilbert_llrd.csv')
distilbert_optimal = pd.read_csv('/content/drive/MyDrive/Dissertation/results/distilbert_optimal.csv')
```

```python
qqplot(bert_results['eval_f1'],line='s' )
pyplot.ylabel("BERT Macro F1-Score")
pyplot.show()
```

```python
qqplot(distilbert_results['eval_f1'],line='s' )
pyplot.ylabel("DistilBERT Macro F1-Score")
pyplot.show()
```

```python
qqplot(distilbert_llrd['eval_f1'],line='s' )
pyplot.ylabel("DistilBERT LLRD Macro F1-Score")
pyplot.show()
```

```python
qqplot(distilbert_optimal['eval_f1'],line='s' )
pyplot.ylabel("DistilBERT WT+LLRD Macro F1-Score")
pyplot.show()
```

```python
print("Shapiro Wilks Test Results")
print("BERT: ", shapiro(bert_results['eval_f1']))
print("DistilBERT: ", shapiro(distilbert_results['eval_f1']))
print("DistilBERT +LLRD : ", shapiro(distilbert_llrd['eval_f1']))
print("DistilBERT +WT + LLRD: ", shapiro(distilbert_optimal['eval_f1']))
```

Scatterplots and Shapiro Wilk tests indicate normality. A large outlier on the DistilBERT +WT+LLRD slightly skewing normality but not significant at alpha 0.05 level. 

```python
print("BERT",ttest_rel(bert_results['eval_f1'], distilbert_results['eval_f1'],alternative="less"))
print("BERT",ttest_rel(bert_results['eval_f1'], distilbert_llrd['eval_f1'],alternative="less"))
print("BERT",ttest_rel(bert_results['eval_f1'], distilbert_optimal['eval_f1'],alternative="less"))
```
